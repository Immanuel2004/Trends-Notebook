```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Year'].fillna(df['Year'].mean(),inplace = True)
```

```python
df['Publisher'].fillna(df['Publisher'].mode(),inplace=True)
```

```python
Global_Sales = df[['Year','Global_Sales']]
```

```python
Global_Sales
```

```python
N_Year = sum(df['Year'])
```

```python
N_Year
```

```python
N_Global_Sales = sum(df['Global_Sales'])
```

```python
N_Global_Sales
```

```python
Global_Sales['X = x - N_year'] = Global_Sales['Year'] + Global_Sales['Global_Sales']
```

```python
Global_Sales
```

```python
Global_Sales['X = x - N_year'] = np.nan
```

```python
Global_Sales
```

```python
len_Year = len(df['Year'])
```

```python
len_Year
```

```python
X = N_Year / len_Year
```

```python
X
```

```python
for i in range(16598):
    New_x = Global_Sales['Global_Sales'].iloc[i] - X  
    Global_Sales.at[i, 'X = x - N_year'] = New_x  
```

```python
Global_Sales
```

```python
Global_Sales['X^2'] = Global_Sales['Year'] + Global_Sales['Global_Sales'] + Global_Sales['X = x - N_year']
```

```python
Global_Sales['X^2'] = np.nan
```

```python
# for i in range(16598):
#     New_x2 = New_x * New_x   
#     Global_Sales.at[i, 'X^2'] = New_x2

for i in range(len(Global_Sales)):
    Global_Sales.at[i, 'X^2'] = Global_Sales['X = x - N_year'].iloc[i] ** 2  

```

```python
Global_Sales
```

```python
Global_Sales['XY'] = Global_Sales['Year'] + Global_Sales['Global_Sales'] + Global_Sales['X = x - N_year'] + Global_Sales['X^2']
Global_Sales['XY'] = np.nan
```

```python
for i in range(16598):
    New_x3 = Global_Sales['Year'].iloc[i] * Global_Sales['Global_Sales'].iloc[i] 
    Global_Sales.at[i, 'XY'] = New_x3  
```

```python
Global_Sales
```

```python
a = sum(Global_Sales['Global_Sales']) / len_Year
```

```python
a
```

```python
b = sum(Global_Sales['XY']) / sum(Global_Sales['X^2'])
```

```python
b
```

```python
Global_Sales['Y = a + bx'] = Global_Sales['Year'] + Global_Sales['Global_Sales'] + Global_Sales['X = x - N_year'] + Global_Sales['X^2'] + Global_Sales['XY']
Global_Sales['Y = a + bx'] = np.nan
```

```python
for i in range(len(Global_Sales)):
    Y = a + b * Global_Sales['X = x - N_year'].iloc[i]  
    Global_Sales.at[i, 'Y = a + bx'] = Y  
```

```python
Global_Sales
```

```python
sns.lineplot(x=Global_Sales['Year'], y=Global_Sales['Y = a + bx'], linestyle='dotted', color='blue', label='y = a + bx')

plt.xlabel('Year')
plt.ylabel('Y = a + bx')
plt.title('Dotted Line Plot of Y = a + bx vs. Year')
plt.legend()
plt.grid(True)
plt.show()
```
